### How can we deal with missing data

1. Drop data 
    1. drop the whole row
    2. drop the whole column
2. Replace data
    1. replace it by mean
    2. replace it by frequency
    3. replace it based on other functions (Fit a model with missing value column as target variable and predict with the          help of other values

```python
#Let's replace the missing values with the mean of the column
```

```python
raw_data.mean()
```

```python
cols = ["normalized-losses","bore","stroke","horsepower","peak-rpm","price"]
```

```python
for i in cols:
    avg = raw_data[i].astype(float).mean(axis=0)
    raw_data[i].replace(np.nan,avg,inplace=True)
    print(avg)
```
